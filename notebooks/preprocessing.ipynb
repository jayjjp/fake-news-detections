```python
import json
from nltk.corpus import stopwords as nltk_stopwords

stopwords_set = set(nltk_stopwords.words('english'))

def parse_and_filter(s):
    # convert string to list
    token_list = json.loads(s.replace("'", '"'))
    # filter stopwords
    return [word for word in token_list if word.lower() not in stopwords_set]

df['filtered_tokens'] = df['text'].apply(parse_and_filter)

print(df['filtered_tokens'].iloc[0][:10])  # sanity check
```

```python
import numpy as np
import pandas as pd
import joblib
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold, cross_val_score
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
from gensim.models import Word2Vec
from xgboost import XGBClassifier
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# ── load full data.csv (not just train split) ─────────────
full_df = pd.read_csv("../data/data.csv")

# ── preprocessing pipeline on full dataset ────────────────
import json, string
from nltk.corpus import stopwords as nltk_stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer

stopwords_set = set(nltk_stopwords.words('english'))
port_stemmer  = PorterStemmer()
word_lem      = WordNetLemmatizer()

def preprocess(text):
    # remove punctuation + lowercase
    text = "".join([c for c in text if c not in string.punctuation]).lower()
    # tokenize
    tokens = text.split()
    # stopword removal
    tokens = [w for w in tokens if w not in stopwords_set]
    # stem + lemmatize
    tokens = [word_lem.lemmatize(port_stemmer.stem(w)) for w in tokens]
    return tokens

# handle both raw strings and already-tokenized strings saved as lists
def safe_preprocess(text):
    if isinstance(text, str) and text.startswith("['"):
        tokens = json.loads(text.replace("'", '"'))
        tokens = [word_lem.lemmatize(port_stemmer.stem(w)) for w in tokens
                  if w not in stopwords_set]
        return tokens
    return preprocess(text)

full_df['lem_words'] = full_df['text'].apply(safe_preprocess)
full_df['joined_text'] = full_df['lem_words'].apply(lambda x: ' '.join(x))

# ── encode labels ──────────────────────────────────────────
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
full_df['label_encoded'] = le.fit_transform(full_df['label'])  # fake=0, real=1
y = full_df['label_encoded'].values
print(f"Dataset size: {len(full_df)} | Classes: {dict(zip(le.classes_, le.transform(le.classes_)))}")

# ── build embeddings on FULL dataset ──────────────────────
# TF-IDF
tfidf = TfidfVectorizer(max_features=5000)
X_tfidf = tfidf.fit_transform(full_df['joined_text'])

# Word2Vec
w2v_model = Word2Vec(sentences=full_df['lem_words'], vector_size=100, window=5, min_count=1, workers=4)
def get_w2v_vector(tokens):
    vectors = [w2v_model.wv[w] for w in tokens if w in w2v_model.wv]
    return np.mean(vectors, axis=0) if vectors else np.zeros(100)
X_w2v = np.array(full_df['lem_words'].apply(get_w2v_vector).tolist())

# save vectorizers for later use
joblib.dump(tfidf, '../models/tfidf_vectorizer.pkl')
w2v_model.save('../models/w2v_model.bin')
print("✅ Vectorizers saved")

# ── experiment tracking ────────────────────────────────────
results        = []
best_score     = 0
best_model     = None
best_model_name    = ''
best_embedding_name = ''

kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

classifier_params = {
    'Logistic Regression': (
        LogisticRegression(max_iter=1000),
        {'C': [0.01, 0.1, 1, 10], 'solver': ['lbfgs', 'liblinear']}
    ),
    'Random Forest': (
        RandomForestClassifier(),
        {'n_estimators': [100, 200], 'max_depth': [None, 10, 20], 'min_samples_split': [2, 5]}
    ),
    'SVM': (
        SVC(kernel='linear'),
        {'C': [0.1, 1, 10]}
    ),
    'Decision Tree': (
        DecisionTreeClassifier(),
        {'max_depth': [None, 10, 20], 'min_samples_split': [2, 5, 10]}
    ),
    'KNN': (
        KNeighborsClassifier(),
        {'n_neighbors': [3, 5, 7, 11], 'weights': ['uniform', 'distance']}
    ),
    'Gradient Boosting': (
        GradientBoostingClassifier(),
        {'n_estimators': [100, 200], 'learning_rate': [0.05, 0.1], 'max_depth': [3, 5]}
    ),
    'XGBoost': (
        XGBClassifier(eval_metric='logloss'),
        {'n_estimators': [100, 200], 'learning_rate': [0.05, 0.1], 'max_depth': [3, 6]}
    ),
    'MLP': (
        MLPClassifier(max_iter=300),
        {'hidden_layer_sizes': [(128,), (256, 128), (256, 128, 64)], 'alpha': [0.0001, 0.001]}
    ),
}

embeddings = {
    'TF-IDF':   X_tfidf,
    'Word2Vec': X_w2v,
}

# ── ML models with GridSearchCV + KFold ───────────────────
for emb_name, X in embeddings.items():
    X_dense = X.toarray() if hasattr(X, 'toarray') else X
    X_train, X_test, y_train, y_test = train_test_split(
        X_dense, y, test_size=0.2, random_state=42, stratify=y
    )

    for clf_name, (clf, param_grid) in classifier_params.items():
        print(f"\n🔍 [{emb_name}] Tuning {clf_name}...")
        try:
            grid = GridSearchCV(clf, param_grid, cv=kfold, scoring='accuracy', n_jobs=-1, verbose=0)
            grid.fit(X_train, y_train)

            best_clf = grid.best_estimator_
            y_pred   = best_clf.predict(X_test)
            acc      = accuracy_score(y_test, y_pred)
            cv_scores = cross_val_score(best_clf, X_dense, y, cv=kfold, scoring='accuracy', n_jobs=-1)

            print(f"✅ {clf_name} | Test: {acc*100:.2f}% | CV: {cv_scores.mean()*100:.2f}% ± {cv_scores.std()*100:.2f}%")
            print(f"   Best Params: {grid.best_params_}")
            print(classification_report(y_test, y_pred, target_names=le.classes_))

            results.append({
                'Embedding': emb_name, 'Model': clf_name,
                'Test Accuracy': round(acc * 100, 2),
                'CV Accuracy':   round(cv_scores.mean() * 100, 2),
                'CV Std':        round(cv_scores.std() * 100, 2),
                'Best Params':   str(grid.best_params_)
            })

            if acc > best_score:
                best_score = acc
                best_model = best_clf
                best_model_name    = clf_name
                best_embedding_name = emb_name

        except Exception as e:
            print(f"❌ [{emb_name}] {clf_name} failed: {e}")

# ── Deep Learning models ───────────────────────────────────
def build_and_train_dl(X_train, X_test, y_train, y_test, arch):
    input_dim = X_train.shape[1]
    layer_map = {
        'DL-128-64':     [128, 64],
        'DL-256-128':    [256, 128],
        'DL-512-256-128':[512, 256, 128]
    }
    layers = layer_map[arch]
    model = Sequential()
    model.add(Dense(layers[0], activation='relu', input_shape=(input_dim,)))
    model.add(Dropout(0.3))
    for units in layers[1:]:
        model.add(Dense(units, activation='relu'))
        model.add(Dropout(0.3))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
    model.fit(X_train, y_train, epochs=30, batch_size=32,
              validation_split=0.1, callbacks=[early_stop], verbose=0)
    _, acc = model.evaluate(X_test, y_test, verbose=0)
    return model, acc

for emb_name, X in embeddings.items():
    X_dense = X.toarray() if hasattr(X, 'toarray') else X
    X_train, X_test, y_train, y_test = train_test_split(
        X_dense, y, test_size=0.2, random_state=42, stratify=y
    )
    for arch in ['DL-128-64', 'DL-256-128', 'DL-512-256-128']:
        print(f"\n🔍 [{emb_name}] Training {arch}...")
        model, acc = build_and_train_dl(X_train, X_test, y_train, y_test, arch)
        print(f"✅ [{emb_name}] {arch} — {acc*100:.2f}%")
        results.append({
            'Embedding': emb_name, 'Model': arch,
            'Test Accuracy': round(acc * 100, 2),
            'CV Accuracy': '-', 'CV Std': '-',
            'Best Params': 'EarlyStopping(patience=3)'
        })
        if acc > best_score:
            best_score = acc
            best_model_name     = arch
            best_embedding_name = emb_name
            model.save('../models/best_dl_model.keras')
            print(f"🏆 New best DL model saved!")

# ── final results ──────────────────────────────────────────
import os
os.makedirs('../models', exist_ok=True)

results_df = pd.DataFrame(results).sort_values('Test Accuracy', ascending=False)
print("\n" + "="*70)
print("FINAL RESULTS".center(70))
print("="*70)
print(results_df[['Embedding', 'Model', 'Test Accuracy', 'CV Accuracy', 'CV Std']].to_string(index=False))

if best_model is not None:
    joblib.dump(best_model, '../models/best_ml_model.pkl')

print(f"\n🏆 Best Model : [{best_embedding_name}] {best_model_name}")
print(f"🎯 Accuracy   : {best_score*100:.2f}%")
print(f"💾 Saved to   : ../models/")
```
